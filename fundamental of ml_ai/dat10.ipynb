```python
import numpy as np
import cv2 
import skimage
import matplotlib.pyplot as plt
```

```python
image1=plt.imread('cat.jpg')
plt.imshow(image1)
plt.axis('off')
```

```python
img_gray=cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)

print(img_gray.shape)
plt.imshow(img_gray,cmap='gray')
```

```python
th,img_binary=cv2.threshold(img_gray, 80,255, cv2.THRESH_BINARY)
img_binary.shape
```

```python
plt.imshow(img_binary,cmap='gray')
```

```python
th,img_binary_rgb=cv2.threshold(image1, 80,255, cv2.THRESH_BINARY)
img_binary_rgb.shape
```

```python
plt.imshow(img_binary_rgb,cmap='gray')
```
